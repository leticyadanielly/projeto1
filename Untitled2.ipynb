```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Caminho da imagem corrigido
caminho_imagem = r"/content/drive/MyDrive/IMG-20160321-WA0004.jpg"

# Abrir a imagem
img = Image.open(caminho_imagem)

# Converter para tons de cinza
img_cinza = img.convert("L")

# Converter a imagem em array NumPy
array_cinza = np.array(img_cinza)

# Definir um limiar para binarização
limiar = 128
array_binarizada = np.where(array_cinza > limiar, 255, 0).astype(np.uint8)

# Criar a imagem binarizada a partir do array
img_binarizada = Image.fromarray(array_binarizada)

# Exibir as três imagens: original, cinza e binarizada
plt.figure(figsize=(15, 5))

# Imagem original
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("Imagem Original")
plt.axis("off")

# Imagem em tons de cinza
plt.subplot(1, 3, 2)
plt.imshow(img_cinza, cmap="gray")
plt.title("Imagem em Tons de Cinza")
plt.axis("off")

# Imagem binarizada
plt.subplot(1, 3, 3)
plt.imshow(img_binarizada, cmap="gray")
plt.title("Imagem Binarizada")
plt.axis("off")

plt.tight_layout()
plt.show()
```
